```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Handle missing data by imputation (filling missing values)
data.fillna(method='ffill', inplace=True)  # Forward fill missing values

# Remove the target variable from the features
X = data.drop('Coal Consumed (UOM:t(Tonnes))', axis=1)
y = data['Coal Consumed (UOM:t(Tonnes))']

# One-hot encode categorical variables
X_encoded = pd.get_dummies(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Train a Random Forest Regressor model
rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor.fit(X_train, y_train)

# Predict on the test set
rf_predictions = rf_regressor.predict(X_test)

# Evaluate the model
rf_mse = mean_squared_error(y_test, rf_predictions)
print("Random Forest Regressor MSE:", rf_mse)

state_wise_predictions = rf_regressor.predict(X_encoded)

# Combine predictions with the original data
predictions_df = pd.DataFrame({'State': data['State'], 'Predicted Coal Consumption (UOM:t(Tonnes))': state_wise_predictions})
print(predictions_df)
```

```python
# Combine predictions with the original data
predictions_df['Year'] = data['Year']  # Adding the 'Year' column from the original data to predictions_df

# Calculate the total coal consumed for each year
total_coal_consumed_actual = data[data['Year'].isin(range(2011, 2019))].groupby('Year')['Coal Consumed (UOM:t(Tonnes))'].sum()
total_coal_consumed_predicted = predictions_df[predictions_df['Year'].isin(range(2018, 2021))].groupby('Year')['Predicted Coal Consumption (UOM:t(Tonnes))'].sum()

# Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(total_coal_consumed_actual.index, total_coal_consumed_actual.values, label='Actual (2011-2018)', marker='o')
plt.plot(total_coal_consumed_predicted.index, total_coal_consumed_predicted.values, label='Predicted (2018-2020)', marker='o')

plt.xlabel('Year')
plt.ylabel('Total Coal Consumed (UOM:t(Tonnes))')
plt.title('Total Coal Consumption Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Combine predictions with the original data
predictions_df['Year'] = data['Year']  # Adding the 'Year' column from the original data to predictions_df

# Calculate the total coal consumed for each year
total_coal_consumed_actual = data[data['Year'].isin(range(2011, 2020))].groupby('Year')['Coal Consumed (UOM:t(Tonnes))'].sum()
total_coal_consumed_predicted = predictions_df[predictions_df['Year'].isin(range(2018, 2021))].groupby('Year')['Predicted Coal Consumption (UOM:t(Tonnes))'].sum()

# Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(total_coal_consumed_actual.index, total_coal_consumed_actual.values, label='Actual (2011-2020)', marker='o')
plt.plot(total_coal_consumed_predicted.index, total_coal_consumed_predicted.values, label='Predicted (2018-2020)', marker='o')

plt.xlabel('Year')
plt.ylabel('Total Coal Consumed (UOM:t(Tonnes))')
plt.title('Total Coal Consumption Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for state in data['State'].unique():
    state_data = data[data['State'] == state]
    plt.plot(state_data['Year'], state_data['Coal Consumed (UOM:t(Tonnes))'], label=f'Original {state} Data')
    predicted_state_data = predictions_df[predictions_df['State'] == state]
    plt.plot(state_data['Year'], predicted_state_data['Predicted Coal Consumption (UOM:t(Tonnes))'], linestyle='dashed', label=f'Predicted {state} Data')

plt.xlabel('Year')
plt.ylabel('Coal Consumed (UOM:t(Tonnes))')
plt.title('Original vs. Predicted Coal Consumption State-wise')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# List of states
states_of_interest = ['Andhra Pradesh', 'Haryana', 'Himachal Pradesh',
                      'Jharkhand', 'Kerala', 'Rajasthan',
                      'Uttar Pradesh']

# Plot line graph for each state
for state in states_of_interest:
    # Filter data for the current state
    state_data = data[data['State'] == state]

    # Calculate the total coal consumed for each year
    total_coal_consumed_actual = state_data[state_data['Year'].isin(range(2011, 2020))].groupby('Year')['Coal Consumed (UOM:t(Tonnes))'].sum()
    total_coal_consumed_predicted = predictions_df[predictions_df['Year'].isin(range(2018, 2021)) & (predictions_df['State'] == state)].groupby('Year')['Predicted Coal Consumption (UOM:t(Tonnes))'].sum()

    # Plot the line graph for the current state
    plt.figure(figsize=(8, 5))
    plt.plot(total_coal_consumed_actual.index, total_coal_consumed_actual.values, label='Actual', marker='o')
    plt.plot(total_coal_consumed_predicted.index, total_coal_consumed_predicted.values, label='Predicted', marker='o')
    plt.xlabel('Year')
    plt.ylabel('Total Coal Consumed (UOM:t(Tonnes))')
    plt.title(f'Total Coal Consumption Over the Years - {state}')
    plt.legend()
    plt.grid(True)
    plt.show()
```
